```python
def ingest_csv_data():
    """Read data from CSV files."""
    try:
        df = pd.read_csv("Dataset/Churn_Modelling.csv")
        logging.info("Successfully ingested churn dataset from CSV.")
        return df
    except Exception as e:
        logging.error(f"Failed to ingest CSV data: {e}")
        return None

def ingest_db_data(table_name):
    """Fetch data from Google BigQuery tables."""
    try:
        query = f"SELECT * FROM `ecommerce-316012.RAW_STG.{table_name}`"
        df = client.query(query).to_dataframe()
        logging.info(f"Successfully ingested {table_name} data from BigQuery.")
        return df
    except Exception as e:
        logging.error(f"Failed to fetch {table_name} data from BigQuery: {e}")
        return None

def save_to_local_storage(df, folder, filename):
    """Save DataFrame to structured local storage with timestamped filenames."""
    try:
        timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
        filename_timestamped = f"{filename.replace('.csv', '')}_{timestamp}.csv"
        if folder == "snapshots":
            folder_path = os.path.join(local_storage_dir, folder)
        else:
            folder_path = os.path.join(local_storage_dir, "csv", folder)
        os.makedirs(folder_path, exist_ok=True)
        file_path = os.path.join(folder_path, filename_timestamped)
        df.to_csv(file_path, index=False)
        logging.info(f"Successfully saved {filename_timestamped} to local storage at {file_path}")
    except Exception as e:
        logging.error(f"Failed to save {filename} to local storage: {e}")
```

```python
def run_ingestion():
    """Main function to run ingestion tasks."""
    churn_data = ingest_csv_data()
    save_to_local_storage(churn_data, "churn_modelling", "churn_data.csv")
    tables = ["SUPPORT_TICKETS", "SATISFACTION_SCORES", "DIGITAL_ENGAGEMENT", "PAYMENT_HISTORY", "COMPETITOR_ANALYSIS"]
    
    db_data = {table: ingest_db_data(table) for table in tables}
    
    if churn_data is not None and all(not df.empty for df in db_data.values() if df is not None):
        
        for table, data in db_data.items():
            if data is not None:
                save_to_local_storage(data, table, f"{table}.csv")
```

```python
import pandas as pd
import os
import warnings
import re
from datetime import datetime

# Suppress warnings
warnings.filterwarnings("ignore")

# Define validation rules based on DDLs

def standardize_geography(df):
    """Standardizes country names and reports invalid ones."""
    country_mapping = {
        "USA": "United States", "US": "United States", "U.K.": "United Kingdom", "UK": "United Kingdom",
        "CAN": "Canada", "GER": "Germany", "FR": "France", "AUS": "Australia", "IND": "India",
        "CN": "China", "JP": "Japan", "BR": "Brazil", "MEX": "Mexico", "SA": "South Africa",
        "IT": "Italy", "ES": "Spain", "NL": "Netherlands", "RU": "Russia", "KR": "South Korea",
        "SE": "Sweden", "NO": "Norway"
    }
    valid_countries = set(country_mapping.values())
    if "Geography" in df.columns:
        invalid_countries = df[~df["Geography"].isin(valid_countries)]["Geography"].unique()
        if len(invalid_countries) > 0:
            return df, f"Invalid country names found: {invalid_countries}"
    return df, None

def contains_special_characters(value):
    """Check if a string contains special characters."""
    return bool(re.search(r'[^a-zA-Z0-9 ]', str(value)))

def validate_data(df, table_name):
    """Perform data validation checks on a DataFrame."""
    report = []
    
    # Expected data types, formats, and ranges
    validation_rules = {
        "CUSTOMER_INFO": {
            "CustomerID": int,
            "Name": str,
            "Age": (int, 18, 100),  # Age should be between 18 and 100
            "Gender": ["Male", "Female"],
            "Geography": str,
            "SubscriptionPlan": ["Basic", "Standard", "Premium"]
        },
        "SUPPORT_TICKETS": {
            "CustomerID": int,
            "CustomerSupportTickets": int,  
            "IssueResolutionTime": int,  
            "ComplaintSeverity": ["Low", "Medium", "High","Critical"],
            "RefundRequests": int
        },
        "SATISFACTION_SCORES": {
            "CustomerID": int,
            "NetPromoterScore": (int, -100, 100),
            "SatisfactionSurveyScore": (int, 1, 10),
            "CustomerLoyaltyProgram": ["Yes", "No"],
            "RewardPointsBalance": int
        },
        "DIGITAL_ENGAGEMENT": {
            "CustomerID": int,
            "MobileAppUsage": int,
            "WebPortalUsage": int,
            "AvgSessionDuration": int,
            "DeviceType": ["Mobile", "Desktop", "Tablet","Smartwatch"]
        },
        "PAYMENT_HISTORY": {
            "CustomerID": int,
            "AutoRenewalEnabled": ["Yes", "No"],
            "PaymentMethod": ["Credit Card", "Debit Card", "PayPal","Bank Transfer","Crypto"],
            "PaymentFailures": int
        },
        "COMPETITOR_ANALYSIS": {
            "CustomerID": int,
            "CompetitorWebsiteVisits": int,
            "ChurnedFriends": int,
            "MarketSentimentScore": ["Positive", "Neutral", "Negative"]
        },
        "CHURN_MODELLING": {
            "CustomerID": int,
            "Surname": str,
            "CreditScore": (int, 300, 850),  # Credit score should be between 300 and 850
            "Geography": str,
            "Gender": ["Male", "Female"],
            "Age": (int, 18, 100),
            "Tenure": (int, 0, 82),
            "Balance": float, 
            "NumOfProducts": int,
            "HasCrCard": [0, 1],
            "IsActiveMember": [0, 1],
            "EstimatedSalary": float,
            "Exited": [0, 1]  # Binary churn indicator
        }
    }
    
    if table_name in validation_rules:
        rules = validation_rules[table_name]
        
        # Check for missing values
        missing_values = df.isnull().sum()
        missing_report = missing_values[missing_values > 0]
        if not missing_report.empty:
            report.append(f"Missing values in {table_name}:\n{missing_report}\n")
        
        # Check for duplicate records
        duplicate_count = df.duplicated().sum()
        if duplicate_count > 0:
            report.append(f"{duplicate_count} duplicate records found in {table_name}.")
        
        # Standardize and validate Geography column if present
        df, geography_issue = standardize_geography(df)
        if geography_issue:
            report.append(geography_issue)
        
        # Check for special characters in string columns
        for col in df.select_dtypes(include=["object"]).columns:
            if df[col].apply(contains_special_characters).any():
                report.append(f"Column {col} in {table_name} contains special characters.")
        
        # Validate data types and ranges
        for col, rule in rules.items():
            if col in df.columns:
                if isinstance(rule, tuple):  # Numeric range check
                    dtype, min_val, max_val = rule
                    df[col] = pd.to_numeric(df[col], errors="coerce")  # Convert non-numeric values to NaN
                    if not df[col].between(min_val, max_val, inclusive="both").all():
                        report.append(f"Column {col} in {table_name} has values out of range [{min_val}, {max_val}].")
                elif isinstance(rule, list):  # Categorical value check
                    if not df[col].isin(rule).all():
                        report.append(f"Column {col} in {table_name} contains invalid values.")
                else:  # Basic data type check
                    if not df[col].map(lambda x: isinstance(x, rule)).all():
                        report.append(f"Column {col} in {table_name} has incorrect data types.")
    
    return report

# Load and validate datasets
validation_results = []
data_folder = "./raw_data/csv/"
tables = ["SUPPORT_TICKETS", "SATISFACTION_SCORES", "DIGITAL_ENGAGEMENT", "PAYMENT_HISTORY", "COMPETITOR_ANALYSIS","churn_modelling"]

for table in tables:
    table_path = os.path.join(data_folder, table.lower())
    if os.path.exists(table_path):
        csv_files = [os.path.join(table_path, f) for f in os.listdir(table_path) if f.endswith('.csv')]
        if not csv_files:
            print(f"No CSV files found in {table_path}, skipping...")
            continue
        latest_file = max(csv_files, key=os.path.getmtime) 
        # latest_file = sorted(os.listdir(table_path))[-1]  # Get the latest file
        df = pd.read_csv(latest_file)
        print(table_path, latest_file)
        validation_results.extend(validate_data(df, table))

# Save validation report
report_path = "./raw_data/validation_reports/"
os.makedirs(report_path, exist_ok=True)
timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
report_file = os.path.join(report_path, f"data_quality_report_{timestamp}.txt")

with open(report_file, "w") as f:
    for result in validation_results:
        f.write(result + "\n")

print(f"Data validation complete. Report saved at: {report_file}")
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import MinMaxScaler

# Load latest merged dataset
merged_cleaned_folder = "./raw_data/cleaned/merged/"
csv_files = [os.path.join(merged_cleaned_folder, f) for f in os.listdir(merged_cleaned_folder) if f.endswith('.csv')]
latest_file = max(csv_files, key=os.path.getmtime)  # Get latest file
print(f"Reading latest merged dataset: {latest_file}")
merged_df = pd.read_csv(latest_file)

# Feature Engineering


# 1. Engagement Score: Sum of digital interactions
merged_df['Engagement_Score'] = merged_df['MobileAppUsage'] + merged_df['WebPortalUsage'] + merged_df['AvgSessionDuration']


# 2. Customer Risk Score: Derived from complaints, failed payments, competitor visits
merged_df['Customer_Risk_Score'] = merged_df['PaymentFailures'] + merged_df['CompetitorWebsiteVisits']

# Drop features used in deriving new ones
columns_to_drop = ['RowNumber', 'MobileAppUsage', 'WebPortalUsage', 'AvgSessionDuration', 'PaymentFailures', 'CompetitorWebsiteVisits']
merged_df.drop(columns=columns_to_drop, inplace=True)

# Apply binning to Age column
merged_df['Age_groups'] = pd.cut(merged_df['Age'], bins=[18, 30, 40, 50, 60, 100], labels=['18-30', '31-40', '41-50', '51-60', '60+'])

# Scale newly created features
scaler = MinMaxScaler()
scaled_features = ['Engagement_Score', 'Customer_Risk_Score']
merged_df[scaled_features] = scaler.fit_transform(merged_df[scaled_features])

print("Feature engineering completed successfully.")



merged_df.columns
```

```python
from google.cloud import bigquery
import pandas as pd

# Define BigQuery settings
PROJECT_ID = "stoked-grin-452213-h1"  # Replace with your GCP project ID
DATASET_ID = "LOAD"  # Replace with your dataset name
TABLE_ID = "CustomerData"  # Target table
TEMP_TABLE_ID = "CustomerData_temp"  # Temporary table for staging

DEST_TABLE = f"{PROJECT_ID}.{DATASET_ID}.{TABLE_ID}"
TEMP_TABLE = f"{PROJECT_ID}.{DATASET_ID}.{TEMP_TABLE_ID}"

# Initialize BigQuery client
client = bigquery.Client()

# Load latest data (merged_df assumed to be prepared in Python)
# merged_df = pd.read_csv("your_merged_file.csv")  # Uncomment if loading from CSV

# Step 1: Upload new data to temporary table
job_config = bigquery.LoadJobConfig(write_disposition="WRITE_TRUNCATE")  # Overwrite temp table
job = client.load_table_from_dataframe(merged_df, TEMP_TABLE, job_config=job_config)
job.result()  # Wait for the job to complete

print(f"Uploaded {len(merged_df)} records to temporary table: {TEMP_TABLE}")

# Step 2: Run MERGE (UPSERT) Query
merge_query = f"""
MERGE `{DEST_TABLE}` AS target
USING `{TEMP_TABLE}` AS source
ON target.CustomerID = source.CustomerID
WHEN MATCHED THEN
  UPDATE SET 
    target.CreditScore = source.CreditScore,
    target.Age_Binned = source.Age_Binned,
    target.Tenure = source.Tenure,
    target.Balance = source.Balance,
    target.NumOfProducts = source.NumOfProducts,
    target.EstimatedSalary = source.EstimatedSalary,
    target.RewardPointsBalance = source.RewardPointsBalance,
    target.PaymentFailures = source.PaymentFailures,
    target.CustomerSupportTickets = source.CustomerSupportTickets,
    target.IssueResolutionTime = source.IssueResolutionTime,
    target.RefundRequests = source.RefundRequests,
    target.MobileAppUsage = source.MobileAppUsage,
    target.WebPortalUsage = source.WebPortalUsage,
    target.AvgSessionDuration = source.AvgSessionDuration,
    target.CompetitorWebsiteVisits = source.CompetitorWebsiteVisits,
    target.ChurnedFriends = source.ChurnedFriends,
    target.Engagement_Score = source.Engagement_Score,
    target.Customer_Risk_Score = source.Customer_Risk_Score,
    target.Exited = source.Exited
WHEN NOT MATCHED THEN
  INSERT (CustomerID, CreditScore, Age_Binned, Tenure, Balance, NumOfProducts, EstimatedSalary,
          RewardPointsBalance, PaymentFailures, CustomerSupportTickets, IssueResolutionTime,
          RefundRequests, MobileAppUsage, WebPortalUsage, AvgSessionDuration, CompetitorWebsiteVisits,
          ChurnedFriends, Engagement_Score, Customer_Risk_Score, Exited)
  VALUES (source.CustomerID, source.CreditScore, source.Age_Binned, source.Tenure, source.Balance, 
          source.NumOfProducts, source.EstimatedSalary, source.RewardPointsBalance, source.PaymentFailures, 
          source.CustomerSupportTickets, source.IssueResolutionTime, source.RefundRequests, source.MobileAppUsage, 
          source.WebPortalUsage, source.AvgSessionDuration, source.CompetitorWebsiteVisits, source.ChurnedFriends, 
          source.Engagement_Score, source.Customer_Risk_Score, source.Exited);
"""

# Execute the MERGE query
query_job = client.query(merge_query)
query_job.result()  # Wait for the job to complete

print(f"MERGE query executed successfully. Data updated in {DEST_TABLE}.")

# Step 3: Drop the temporary table
drop_query = f"DROP TABLE `{TEMP_TABLE}`"
client.query(drop_query)

print(f"Temporary table {TEMP_TABLE} dropped.")

```

## Data Versioning Strategy and Workflow

### Overview
we use **DVC (Data Version Control)** and **Git** to track and manage dataset versions. The versioning workflow ensures that data changes are reproducible, traceable, and securely stored in remote storage.

### Workflow
1. **Check for Dataset Existence**  
   - The script verifies if the dataset (`merged_output.csv`) exists before proceeding.

2. **Add Dataset to DVC Tracking**  
   - `dvc add merged_output.csv` registers the dataset in DVC.
   - This generates a `.dvc` file that records the dataset's version metadata.

3. **Commit DVC Metadata to Git**  
   - The `.dvc` file and `.gitignore` are staged and committed to Git.
   - This ensures that dataset versions are linked with Git commits.

4. **Push Dataset to DVC Remote Storage**  
   - `dvc push` uploads the dataset to the remote storage configured in DVC.

5. **Store Version Metadata**  
   - The script captures the latest Git commit hash and timestamps the dataset version.
   - The metadata is stored in `data_version_metadata.json`.

6. **Commit and Push Metadata**  
   - The version metadata file is added, committed, and pushed to Git.
   - This keeps a record of dataset versions alongside code changes.

### Benefits
✅ **Reproducibility** - Enables recreating previous dataset versions.  
✅ **Collaboration** - Ensures teams work with the same dataset version.  
✅ **Traceability** - Maintains a log of dataset modifications over time.  
✅ **Storage Efficiency** - Prevents redundant storage of large datasets in Git.  

With this setup, dataset changes are versioned systematically, ensuring a robust and scalable workflow for Machine Learning projects.
